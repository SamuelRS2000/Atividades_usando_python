# MULTA(R$)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dados = pd.read_excel('DADOS.xlsx', engine='openpyxl')

multa_r = dados['Multa'].to_numpy()
datas = dados['Data'].to_numpy()
datas_a = datas.astype('str')

xs = [f'Setembro/{a[8:10]}' for a in datas_a if a[5] == '0'] #Quando a 5 letre astring for 0 , e setembro 
xo = [f'Outubro/{a[8:10]}' for a in datas_a if a[5] == '1']  #Quando a 5 letre astring for 1 , e outubro 

xo, xs = np.unique(xo), np.unique(xs)
x = np.append(xs, xo)


#----------------------------PLOTAGEM
font1 = {'family': 'serif', 'color': 'black', 'size': 40}
font2 = {'family': 'serif', 'color': 'red', 'size': 40}

plt.figure(figsize=(20, 10)) #tamanho 
plt.plot(datas, multa_r,'blue',alpha = 0.8) #tipo do plot 
plt.title("Multa por Excesso de Reativo (R$)", fontdict=font1,fontweight='bold',pad=30) #titulo 
plt.grid(axis='y')           #grade

# Definir manualmente os ticks no eixo y
plt.yticks(np.arange(0, 2.2, 0.25), [f'R${round(val, 2)}' for val in np.arange(0, 2.1, 0.25)],color = 'red',fontdict=font1,fontweight='bold')

# Definir manualmente os ticks no eixo x
plt.xticks(np.linspace(18521.3,18548.3,28), x, rotation=45,fontdict=font1,fontweight='bold')

#colocar o valor em pontos especificos
i = 0 
for xitem,yitem in np.nditer([datas,multa_r]):
    etiqueta = f'{np.around(yitem,2)}'
    i+=1
    if i>20 : 
        plt.annotate(etiqueta, (xitem,yitem), textcoords="offset points",xytext=(0,20),ha="center",fontweight='bold')
        plt.plot(xitem,yitem,'ro')
        i = -90
plt.legend(['Valor'],fontsize = 13) #Legenda
xmin, xmax, ymin, ymax = plt.axis([18521.3,18549.5,0,2]) #limite do eixos
plt.tick_params(labelsize=15) #tamanho dos eixos

plt.show()
```
